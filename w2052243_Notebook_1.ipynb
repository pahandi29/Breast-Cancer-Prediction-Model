Code reuse session 1 Part 23

```python
# Encode categorical features to numeric

df['Mortality_Status'] = df['Mortality_Status'].map({'Dead': 0, 'Alive': 1})
df['T_Stage'] = df['T_Stage'].map({'T1': 0, 'T2': 1, 'T3': 2, 'T4': 3})
df['N_Stage'] = df['N_Stage'].map({'N1': 0, 'N2': 1, 'N3': 2})
df['6th_Stage'] = df['6th_Stage'].map({'IIA': 0, 'IIIA': 1, 'IIB': 2, 'IIIB': 3, 'IIIC':4})
df['Differentiated'] = df['Differentiated'].map({
    'Poorly differentiated': 0,
    'Moderately differentiated': 1,
    'Well differentiated': 2,
    'Undifferentiated': 3
})
df['A_Stage'] = df['A_Stage'].map({'Regional': 0, 'Distant': 1})
df['Estrogen_Status'] = df['Estrogen_Status'].map({'Negative': 0, 'Positive': 1})
df['Progesterone_Status'] = df['Progesterone_Status'].map({'Negative': 0, 'Positive': 1})

# Check encoding
df[['T_Stage', 'N_Stage', 'Mortality_Status']].head()
```

Code reuse session 1 Part 17

```python
# Handle outliers in Age using IQR
Q1_age, Q3_age = df['Age'].quantile([0.25, 0.75])
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
df = df[(df['Age'] >= lower_bound_age) & (df['Age'] <= upper_bound_age)]

# Handle outliers in Tumor_Size (only lower fence)
Q1_tumor, Q3_tumor = df['Tumor_Size'].quantile([0.25, 0.75])
IQR_tumor = Q3_tumor - Q1_tumor
lower_bound_tumor = Q1_tumor - 1.5 * IQR_tumor
upper_bound_tumor  = Q3_tumor + 1.5 * IQR_tumor
df = df[(df['Tumor_Size'] >= lower_bound_tumor) & (df['Tumor_Size'] <= upper_bound_tumor)]

print("Shape after outlier handling:", df.shape)
```
